### Machine Learning (Decision Tree);

```python
import numpy as np
```

```python
inputs=df.drop('deposit',axis="columns") 
target=df['deposit']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
a=LabelEncoder()
```

```python
df.columns
```

```python
inputs['job1']  = a.fit_transform(inputs['job'])
inputs['marital1']  =  a.fit_transform(inputs['marital']) 
inputs['education1'] = a.fit_transform(inputs['education']) 
inputs['default1'] = a.fit_transform(inputs['default']) 
inputs['housing1'] = a.fit_transform(inputs['housing']) 
inputs['loan1'] = a.fit_transform(inputs['loan']) 
inputs['month1'] = a.fit_transform(inputs['month']) 
inputs['poutcome1'] = a.fit_transform(inputs['poutcome']) 
```

```python
inputs.head()
```

```python
inputs1=inputs.drop(['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'duration', 'campaign', 'pdays', 'poutcome'
],
axis="columns") 
inputs1.head()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(inputs1,target,test_size=0.2,random_state=3)
```

```python
print(X_train.shape) 
print(X_test.shape)
```

```python
print(Y_train.shape) 
print(Y_test.shape)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
Y_train.head()
```

```python
Y_test.head()
```

```python
model  =  tree.DecisionTreeClassifier()
model.fit(inputs1,target)
```

```python
model.score(inputs1,target)
```

```python
from sklearn.metrics import accuracy_score
```

```python
prediction_on_training_data = model.predict(X_train)
accuracy_on_training_data = accuracy_score(Y_train,prediction_on_training_data)
```

```python
print("Accuracy on training data: ",accuracy_on_training_data)
```

```python
prediction_on_test_data=model.predict(X_test) 
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
```

```python
print("Accuracy on testing data: ",accuracy_on_test_data)
```
